# Exploring Activation Functions in Machine Learning


Activation functions are crucial components of artificial neural networks for their role in determining whether a neuron should be activated or not. Plus, they introduce non-linearity to neural networks, enabling them to learn complex patterns and relationships within the data.

The choice of activation function can significantly impact the model's performance and training speed. Hence, in this notebook, we'll be diving into various activation functions in Machine Learning using Python.

To start off, let us create a variable `x` which will serve as the input coming to our activation functions.

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(42) # Seeding all values

x = np.linspace(-5, 5, 100)
print(x)
```

Next, creating a helper function to plot our activation functions.

```python
def plot_act_func(x_axis_vals, y_axis_vals, title_str):
    plt.figure(figsize=(8, 4))
    plt.plot(x_axis_vals, y_axis_vals, label=title_str.split(" ")[0], color='blue')
    plt.title(title_str)
    plt.xlabel('Input')
    plt.ylabel('Output')
    plt.legend()
    plt.grid(True)
    plt.show()
```

From this point forward, we'll delve into all the contemporary activation functions commonly employed in modern-day Machine Learning.

- Sigmoid Function

$$ \sigma(x) = \frac{1} {1 + e^{-x}} $$

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
    
sigmoid_values = sigmoid(x)

plot_act_func(x, sigmoid_values, title_str="Sigmoid Activation Function")
```

- tanh Function

$$ tanh(x) = \frac{e^x - e^{-x}}{e^x + e^{-x}} = \frac{1 - e^{-2x}}{1 + e^{-2x}} $$

```python
def tanh(x):
    return (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))

tanh_values = tanh(x)

plot_act_func(x, tanh_values, title_str="Hyperbolic-Tangent Activation Function")
```

- Rectified Linear Unit (ReLU) Function

$$ ReLU(x) = max(0, x) $$

```python
def relu(x):
    return np.maximum(0, x)

relu_values = relu(x)

plot_act_func(x, relu_values, title_str="ReLU Activation Function")
```

- Leaky Rectified Linear Unit (LeakyReLU) function

$$
\begin{equation} f(x) = \begin{cases} αx, & \text{if}\ x < 0 \\ x, & \text{otherwise} \\ \end{cases} \end{equation}
$$


```python
def leaky_relu(x, alpha=0.01):
    return np.where(x > 0, x, alpha * x)

leaky_relu_values = leaky_relu(x)

plot_act_func(x, leaky_relu_values, title_str="Leaky-ReLU Activation Function")
```

- Parametric Rectified Linear Unit (PReLU) Function

$$
f\left(x_{i}\right) = x_{i} \text{ if } x_{i} \ge 0 \\

f\left(x_{i}\right) = a_{i}x_{i} \text{ if } x_{i} \leq 0
$$

```python
class PReLU:
    def __init__(self, alpha=0.01):
        self.alpha = alpha

    def __call__(self, x):
        return np.where(x > 0, x, self.alpha * x)
```

```python
for alpha_value in [0.1, 0.01, 0.001]:

    prelu = PReLU(alpha=alpha_value)
    prelu_values = prelu(x)

    plot_act_func(x, prelu_values, title_str=f"PReLU-α={alpha_value} Activation Function")
```

- Exponential Linear Unit (ELU) Function

$$
f\left(x\right) = x \text{ if } x > 0 \\ 

\alpha\left(\exp\left(x\right) − 1\right) \text{ if } x \leq 0
$$


```python
def elu(x, alpha=1.0):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))

elu_values = elu(x)

plot_act_func(x, elu_values, title_str="ELU Activation Function")
```

- Scaled Exponential Linear Unit (SELU) Function

$$
f\left(x\right) = \lambda{x} \text{ if } x \geq{0} \\
f\left(x\right) = \lambda{\alpha\left(\exp\left(x\right) -1 \right)} \text{ if } x < 0
$$

```python
def selu(x, alpha=1.67326, scale=1.0507):
    return scale * np.where(x > 0, x, alpha * (np.exp(x) - 1))

selu_values = selu(x)

plot_act_func(x, selu_values, title_str="SELU Activation Function")
```

- Swish function

$$
f(x) = x \cdot \text{sigmoid}(\beta x)
$$

```python
def swish(x, beta=1.0):
    return x / (1 + np.exp(-beta * x))

swish_values = swish(x)

plot_act_func(x, swish_values, title_str="Swish Activation Function")
```

- Gaussian Error Linear Unit (GELU) Function

$$ \text{GELU}\left(x\right) = x{P}\left(X\leq{x}\right) = x\Phi\left(x\right) = x \cdot \frac{1}{2}\left[1 + \text{erf}(x/\sqrt{2})\right], $$

And, its approximated form is,

$$
\text{GELU}\left(x\right) = 0.5x\left(1+\tanh\left[\sqrt{2/\pi}\left(x + 0.044715x^{3}\right)\right]\right)
$$

```python
def gelu(x):
    return x * 0.5 * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3))))

gelu_values = gelu(x)

plot_act_func(x, gelu_values, title_str="GELU Activation Function")
```

- Softmax Function

$$ \sigma(x_i) = \frac{e^{x_{i}}}{\sum_{j=1}^K e^{x_{j}}} \ \ \ for\ i=1,2,\dots,K $$

```python
def softmax(x):
    exp_x = np.exp(x - np.max(x))  # for numerical stability
    return exp_x / np.sum(exp_x)

softmax_values = softmax(x)

plot_act_func(x, softmax_values, title_str="Softmax Activation Function")
```

# Linear-Transformation based Activation Functions

- Gated Linear Unit (GLU) Function

$$
\text{GLU}\left(a, b\right) = a\otimes \sigma\left(b\right)
$$

```python
def glu(x, gating):
    return np.multiply(x, sigmoid(gating))

x = np.array([[1, 2, 3],
              [4, 5, 6]])

gating = np.array([[0.6, 0.2, 0.3],
                   [0.3, 0.1, 0.2]])

glu_values = glu(x, gating=gating)

print(glu_values)
```

- ReGLU (ReLU + GLU) Function

$$
\text{ReGLU}\left(x, W, V, b, c\right) = \max\left(0, xW + b\right) \otimes \left(xV + c\right)
$$

```python
def relu(x):
    return np.maximum(0, x)

def gated_linear_unit(x):
    return sigmoid(x)

def reglu(x, W, V, b, c):
    relu_output = relu(np.dot(x, W) + b)
    glu_output = gated_linear_unit(np.dot(x, V) + c)
    return np.multiply(relu_output, glu_output)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Example usage
x = np.array([[1, 2, 3],
              [4, 5, 6]])

W = np.random.randn(3, 3)
V = np.random.randn(3, 3)
b = np.random.randn(3)
c = np.random.randn(3)

output = reglu(x, W, V, b, c)
print(output)
```

- GeGLU (GELU + GLU) Function

$$
\text{GeGLU}\left(x, W, V, b, c\right) = \text{GELU}\left(xW + b\right) \otimes \left(xV + c\right)
$$

```python
def gelu(x):
    return 0.5 * x * (1 + np.tanh(np.sqrt(2 / np.pi) * (x + 0.044715 * np.power(x, 3))))

def gated_linear_unit(x):
    return sigmoid(x)

def geglu(x, W, V, b, c):
    gelu_output = gelu(np.dot(x, W) + b)
    glu_output = gated_linear_unit(np.dot(x, V) + c)
    return np.multiply(gelu_output, glu_output)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Example usage
x = np.array([[1, 2, 3],
              [4, 5, 6]])

W = np.random.randn(3, 3)
V = np.random.randn(3, 3)
b = np.random.randn(3)
c = np.random.randn(3)

output = geglu(x, W, V, b, c)
print(output)
```

- SwiGLU (Swish + GLU) Function

$$
\text{SwiGLU}\left(x, W, V, b, c, \beta\right) = \text{Swish}_{\beta}\left(xW + b\right) \otimes \left(xV + c\right)
$$

```python
def swish(x, beta=1.0):
    return x * sigmoid(beta * x)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def gated_linear_unit(x):
    return sigmoid(x)

def swiglu(x, W, V, b, c, beta=1.0):
    swish_output = swish(np.dot(x, W) + b, beta)
    glu_output = gated_linear_unit(np.dot(x, V) + c)
    return np.multiply(swish_output, glu_output)

x = np.array([[1, 2, 3],
              [4, 5, 6]])

W = np.random.randn(3, 3)
V = np.random.randn(3, 3)
b = np.random.randn(3)
c = np.random.randn(3)
beta = 1.0

output = swiglu(x, W, V, b, c, beta)
print(output)
```
